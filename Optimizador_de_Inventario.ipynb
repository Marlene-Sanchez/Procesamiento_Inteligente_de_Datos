```python
# Formato: (ID_Producto, Categoría, Precio, Stock_Actual)
entradas_almacen = [
    ("PROD-001", "Electrónica", 150.0, 5),
    ("PROD-002", "Hogar", 25.0, 10),
    ("PROD-001", "Electrónica", 150.0, 5),  # Duplicado
    ("PROD-003", "Ropa", 12.0, 0),
    ("PROD-004", "Electrónica", 300.0, 2),
    ("PROD-002", "Hogar", 25.0, 10),        # Duplicado
    ("PROD-005", "Ropa", 45.0, 15)
]
```

**¿Por qué podemos convertir una lista de tuplas a un set sin errores, pero no podríamos hacerlo si fuera una lista de listas? (Pista: Inmutabilidad).**

 Podemos convertir una lista de tuplas a un set porque las tuplas son inmutables y tienen un valor de identidad único que no cambia. No podríamos hacerlo con una lista de listas porque las listas son mutables; el sistema no puede garantizar que el contenido de la lista no cambie después de haberla guardado en el set, lo que rompería la estructura de datos.

```python
ubicacion_almacen = (19.4326, -99.1332)

try:
    ubicacion_almacen[0] = 20.0
except TypeError as e:
    print(f"Error detectado: {e}")

# Las tuplas protegen la integridad de los datos porque, una vez creadas,
# no pueden ser modificadas (son de solo lectura). Esto evita que un error
# en el código mueva accidentalmente la ubicación del almacén en el sistema.
```

```python
pedidos_urgentes = []
for producto in productos_unicos:
    if producto[3] == 0:
        pedidos_urgentes.append(producto[0])

print(f"Productos urgentes: {pedidos_urgentes}")
```

```python
resumen_almacen = {
    "total_items": len(productos_unicos),
    "categorias": {p[1] for p in productos_unicos},
    "alerta_stock": len(pedidos_urgentes) > 0
}

print("\n3. Resumen del Almacén:")
print(resumen_almacen)
```
